# 2)

Variação menor ou igual a 15% → baixa dispersão: dados homogêneos

Variação entre 15 e 30% → média dispersão

Variação maior que 30% → alta dispersão: dados heterogêneos


Fiz uma conta para pegar o coeficiente de variação
Dei std na minha variavel que contem o total de anos(fiz o mesmo para mes e semana) para pega o desvio padrão

Usei mean da mesma variavel e dividi o std pelo mean e multipliquei por 100 para ter a % e assim temos a variação
em porcentagem

```python
cfanos = (anos.std()/anos.mean())*100
print('Variação no periodo de anos: %.2f' %cfanos)
```

Vemos acima que durante os anos vario bastante a quantidade justamente porque em 2017 tivemos muito poucos emails
comparado aos outros anos

```python
mes.std()
mes.mean()
cfmes = (mes.std()/mes.mean())*100
```

```python
print('Variação no periodo de mês: %.2f' %cfmes)
```

Ja no mes tivemos uma variação quase que média

```python
semana.std()
semana.mean()
cfsemana = (semana.std()/semana.mean())*100
print('Variação no periodo de semana: %.2f' %cfsemana)
```

E nos dias tivemos uma variação menor porem média de fato

# 4)

Utilizei uma função que o professor uso antes na semana 7 para criar a media da hora que vai pega de 1 hora em 
1 hora  de todos os anos ao mesmo tempo

```python
def media_hora(x, nc=24):
    fa,classes= np.histogram(x, bins=nc) 
    fa = fa.astype(int)
    classe_str = []
    for i in range(len(classes)):
        if i<=len(classes)-2:
            classe_str.append('{:2.2f} |-- {:2.2f}' .format(classes[i],classes[i+1]))
        else:
            break        
    return pd.DataFrame({'Horas':classe_str, 'Média de Emails': fa}).set_index('Horas')
```

Chamei a função e guardei ela na variavel e logo após fiz o grafico para visualização

```python
hora = media_hora(df['timeofday'], nc=24)
```

Como podemos ver o horario que mais recebo emails é de 4 as 5 da manhã e também de 9 as 10 horas da manhã

```python
hora.plot.barh(color='red')
```

```python
text = ' '.join(map(str, df.values))
```

Criamos a nuvem e como podemos ver o nome que mais apareceu foi o meu nome completo 

```python
stopwords = ['Re', 'Fwd', '3A_','UTF','Q','gmail',"com'", "inbox'", '2Epng_', '2Ejpg_']
wrd = WordCloud(width=700, height=480, margin=0, collocations=False, background_color="#1F1F1F")
for sw in stopwords:
    wrd.stopwords.add(sw)
wordcloud = wrd.generate(text)
plt.figure(figsize=(25,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
```
