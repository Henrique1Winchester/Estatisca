# 2)

Variação menor ou igual a 15% → baixa dispersão: dados homogêneos

Variação entre 15 e 30% → média dispersão

Variação maior que 30% → alta dispersão: dados heterogêneos

```python
anos.std()
anos.mean()
cfanos = (anos.std()/anos.mean())*100
print('Variação no perido de anos: %.2f' %cfanos)
```

```python
mes.std()
mes.mean()
cfmes = (mes.std()/mes.mean())*100
```

```python
print('Variação no perido de mês: %.2f' %cfmes)
```

```python
semana.std()
semana.mean()
cfsemana = (semana.std()/semana.mean())*100
print('Variação no perido de semana: %.2f' %cfsemana)
```

# 4)

```python
def media_hora(x, nc=24):
    fa,classes= np.histogram(x, bins=nc) 
    fa = fa.astype(int)
    classe_str = []
    for i in range(len(classes)):
        if i<=len(classes)-2:
            classe_str.append('{:2.2f} |-- {:2.2f}' .format(classes[i],classes[i+1]))
        else:
            break        
    return pd.DataFrame({'Horas':classe_str, 'Média de Emails': fa}).set_index('Horas')
```

```python
hora = media_hora(df['timeofday'], nc=24)
```

```python
hora.plot.barh(color='red')
```

```python
text = ' '.join(map(str, df.values))
```

```python
stopwords = ['Re', 'Fwd', '3A_','UTF','Q','gmail',"com'", "inbox'", '2Epng_', '2Ejpg_']
wrd = WordCloud(width=700, height=480, margin=0, collocations=False, background_color="#1F1F1F")
for sw in stopwords:
    wrd.stopwords.add(sw)
wordcloud = wrd.generate(text)
plt.figure(figsize=(25,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
```
